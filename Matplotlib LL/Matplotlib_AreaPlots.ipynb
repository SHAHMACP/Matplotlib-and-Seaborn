# Filled area chart using plotly in Python

```python
import plotly.express as px
import numpy as np

# creating random data through randomint
# function of numpy.random
np.random.seed(42)

random_x= np.random.randint(1,101,100)
random_y= np.random.randint(1,101,100)

fig = px.area(x = random_x, y = random_y)
fig.show()
```

```python
import pandas as pd

data = {'Year': [2015, 2016, 2017, 2018, 2019],'Sales': [100, 150, 200, 180, 250]}

df = pd.DataFrame(data)

import plotly.express as px

fig = px.area(df, x='Year', y='Sales')

fig.show()
```

```python
import pandas as pd

import plotly.express as px

# Create a dataframe with random values
data = pd.DataFrame({
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Category A': [10, 20, 30, 40, 50],
    'Category B': [20, 30, 40, 50, 60],
    'Category C': [30, 40, 50, 60, 70]

})

# Create a stacked area chart

fig = px.area(data, x='Year', y=['Category A', 'Category B', 'Category C'],
              title='Stacked Area Chart')

fig.show()
```

```python
import pandas as pd

import plotly.express as px

# Create a dataframe with random values

data = pd.DataFrame({
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Category A': [10, 20, 30, 40, 50],
    'Category B': [20, 30, 40, 50, 60],
    'Category C': [30, 40, 50, 60, 70]

})

# Create a stacked area chart

fig = px.area(data, x='Year', y=['Category A', 'Category B', 'Category C'],
              title='Stacked Area Chart',
              color_discrete_sequence=['#000000', '#FFFF00', '#800000'])

fig.show()
```

```python
import plotly.express as px

df = px.data.iris()

fig = px.area(df, x="sepal_width", y="sepal_length",
			color="species",
			hover_data=['petal_width'],)

fig.show()
```
