# **Bar Plot**

> Mainly, the barplot shows the relationship between the numeric and categoric values.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

```python
# Sample Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 30, 10, 25, 20]
```

```python
# Basic Bar Plot (Default settings)
plt.figure()
plt.bar(categories, values)
plt.title("Basic Bar Plot")
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

```python
#define array
data= [5, 25, 50, 20]
plt.bar(range(len(data)), data,color='c')
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

```python
# Custom Figure Size & Background
plt.figure(figsize=(8, 6), facecolor='lightgray')
plt.bar(categories, values)
plt.title("Custom Figure Size & Background")
plt.show()
```

```python
# Bar Color & Edge Customization
colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A6', '#FFD433']
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5)
plt.title("Custom Bar Colors & Edges")
plt.show()
```

```python
# Title & Labels Styling
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5)
plt.title("Styled Title & Labels", fontsize=16, fontweight='bold', color='darkblue', pad=10)
plt.xlabel("Categories", fontsize=14, fontweight='bold', color='darkgreen', labelpad=10)
plt.ylabel("Values", fontsize=14, fontweight='bold', color='darkred', labelpad=10)
plt.show()
```

```python
# Adding Grid Customization
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5)
plt.grid(axis='y', linestyle='--', alpha=0.6, color='gray', linewidth=0.7)
plt.title("Grid Customization")
plt.show()
```

```python
# Custom Tick Marks & Rotation
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5)
plt.xticks(fontsize=12, fontweight='bold', rotation=45, color='darkblue')
plt.yticks(fontsize=12, fontweight='bold', color='darkred')
plt.title("Custom Tick Marks & Rotation")
plt.show()
```

```python
# Spines (Border) Customization
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5)
plt.gca().spines['top'].set_visible(False)   #gca - get current axes (to modify axes)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_color('gray')
plt.gca().spines['bottom'].set_color('yellow')
plt.title("Spines (Border) Customization")
plt.show()
```

```python
### Adding a Legend
plt.figure(figsize=(8, 6))
bars = plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5, width=0.6, label="Bar Values")
plt.legend(loc='upper right', fontsize=12, edgecolor='black', title="Legend", title_fontsize=14, shadow=True, framealpha=1, fancybox=True)
plt.title("Legend Customization")
plt.show()
```

```python
x= np.array(['a','b','c','d'])
y= np.array([3,8,7,4])
plt.bar(x,y,color=['r','g','y','b'],edgecolor='k',width=2,lw=3)
plt.show()
```

```python
x= np.array(['a','b','c','d'])
y= np.array([3,8,7,4])
plt.barh(x,y,color=['r','w','y','b'],edgecolor='k',lw=4,height=2)#height is to increase the width of the horizontal bar
plt.grid()
plt.show()
```

# **Unstacked Bar Plots**
Unstacked bar plots compare a specific category over time using different samples.


```python
import pandas as pd
plotdata = pd.DataFrame({

    "2018":[57,67,77,83],

    "2019":[68,73,80,79],

    "2020":[73,78,80,85]},

    index=["Django", "Gafur", "Tommy", "Ronnie"])

plotdata.plot(kind="bar",figsize=(15, 8))
plt.title("FIFA ratings")
plt.xlabel("Footballer")
plt.ylabel("Ratings")
```

```python
import pandas as pd
plotdata = pd.DataFrame({
    "2018":[57,67,77,83],
    "2019":[68,73,80,79],
    "2020":[73,78,80,85]},
    index=["Django", "Gafur", "Tommy", "Ronnie"])
plotdata.plot(kind='bar', stacked=True,figsize=(15, 8))
plt.title("FIFA ratings")

plt.xlabel("Footballer")

plt.ylabel("Ratings")
```

```python
# Example time series data
time = ['2020-01-01', '2020-01-02', '2020-01-03', '2020-01-04']
data = [100, 120, 130, 140]

# Create a pandas dataframe from the time series data
df = pd.DataFrame({'Time': time, 'Data': data})

# Plot the time series data as a bar plot
plt.bar(df['Time'], df['Data'], width=0.8)
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
categories = ['A', 'B', 'C', 'D']
values1 = [5, 7, 3, 8]
values2 = [6, 5, 8, 4]

# Create Bar Plot
fig, ax = plt.subplots(figsize=(8, 5))

# Plot two sets of bars
bar_width = 0.4
x = np.arange(len(categories))

ax.bar(x - bar_width/2, values1, width=bar_width, label="Dataset 1", color='b', alpha=0.7)
ax.bar(x + bar_width/2, values2, width=bar_width, label="Dataset 2", color='r', alpha=0.7)

# Customize Legend
ax.legend(
    loc='upper right',     # Change to 'upper left', 'lower right', etc.
    fontsize=12,
    facecolor='lightgray', # Background color
    edgecolor='black',
    framealpha=0.8         # Transparency
)

# Labels and Title
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.set_title("Bar Plot with Customized Legend")
ax.set_xlabel("Categories")
ax.set_ylabel("Values")

plt.show()
```
