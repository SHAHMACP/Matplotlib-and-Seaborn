> **matplotlib.pyplot.subplots()** is a powerful function in Python that allows users to create a grid of subplots within a single figure.

> With subplots(), users can easily create multiple plots in a single figure, making comparing and analyzing different datasets or variables easier.

> Users can specify the number of rows and columns in the subplot grid and the size and spacing between each subplot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sin(x)')
axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cos(x)')
axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tan(x)')
axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exp(x)')

plt.tight_layout()
plt.show()
```

`fig, axes = plt.subplots(2, 2)`

The function returns two objects:
- fig, which means the entire figure
- axes, which is an array of axes objects representing each subplot.

The Axes objects can be used to modify the properties of each subplot, such as the title, labels, and data.

## spacing between subplots

* The `wspace` parameter controls the horizontal spacing between subplots
* The `hspace` parameter controls the vertical spacing.

```python
fig, axs = plt.subplots(2, 1, figsize=(6, 6))
fig.subplots_adjust(wspace=1, hspace=5)
# Plotting code for the subplots
plt.show()
```

## titles and axis labels of subplots

use the `set_title()` function to add titles to subplots.

To add axis labels to subplots, we can use the `set_xlabel()` and `set_ylabel()` functions.

```python
# Generating sample data
x = np.arange(1, 6)
y1 = np.random.randint(1, 10, size=5)
y2 = np.random.randint(1, 10, size=5)
y3 = np.random.randint(1, 10, size=5)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

ax1.plot(x, y1)
ax1.set_title("Plot 1")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")

ax2.scatter(x, y2)
ax2.set_title("Plot 2")
ax2.set_xlabel("X-axis")
ax2.set_ylabel("Y-axis")

ax3.bar(x, y3)
ax3.set_title("Plot 3")
ax3.set_xlabel("X-axis")
ax3.set_ylabel("Y-axis")

# Plotting code for the subplots
plt.show()
```

```python
x = np.linspace(0., 3*np.pi, 100) # 0 to 3*Pi in 100 steps
y_1 = np.sin(x)
y_2 = np.cos(x)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(9, 5),
                         sharey='row', sharex='row')
axes[0, 0].plot(x+1, y_1+1, '-', c='orange')
axes[0, 1].plot(x, y_2, '-', c='magenta')
axes[1, 0].plot(x, y_1**2, '--', c='orange')
axes[1, 1].plot(x, y_2**2, '--', c='magenta')
axes[0, 0].set_ylabel(r'sin(x)')
axes[0, 1].set_ylabel(r'cos(x)')
axes[1, 0].set_ylabel(r'sin$^2$(x)')
axes[1, 1].set_ylabel(r'cos$^2$(x)')
fig.tight_layout()
```

```python
x = np.linspace(0., 3*np.pi, 100) # 0 to 3*Pi in 100 steps
y_1 = np.sin(x)
y_2 = np.cos(x)

fig = plt.figure(figsize=(8, 6))

ax1 = plt.subplot(2, 2, 1, frameon=True)
ax1.plot(x+1, y_1+1)
ax1.set_title('ax1')

ax2 = plt.subplot(2, 2, 2, sharex=ax1, facecolor='orange')
ax2.plot(x, y_2, '-r')
ax2.set_title('Shares x-axis with ax1')

ax3 = plt.subplot(2, 2, 3, sharey=ax1)
ax3.plot(x, y_1**2, '-g')
ax3.set_title('Shares y-axis with ax1')

ax4 = plt.subplot(2, 2, 4, facecolor='orchid')
ax4.plot(x, y_2**2, '-b')
fig.tight_layout()
```

```python
# Generating sample data
x = np.arange(1, 6)
y = np.random.randint(1, 10, size=5)

def add_title(axes):
    for i, ax in enumerate(axes):
        ax.set_title("ax%d" % (i+1), fontsize=18)

fig = plt.figure(figsize=(8, 8))

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
ax1.plot(x,y)

ax2 = plt.subplot2grid((3, 3), (0, 2), rowspan=3)
ax2.bar(x,y)

ax3 = plt.subplot2grid((3, 3), (1, 0), rowspan=2)
ax3.scatter(x,y)

ax4 = plt.subplot2grid((3, 3), (1, 1))
ax4.pie(y)

ax5 = plt.subplot2grid((3, 3), (2, 1))
ax5.barh(x,y)

add_title(fig.axes)

fig.tight_layout()
```

```python


import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig = plt.figure(figsize=(10, 6))

# Position 1: Full-width on top
ax1 = plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st position
ax1.plot(x, y, 'r')
ax1.set_title("Position (2,2,1)")

# Position 2: Second column, first row
ax2 = plt.subplot(2, 2, 2)  # 2nd position
ax2.scatter(x, np.cos(x), color='b')
ax2.set_title("Position (2,2,2)")

# Position 3: Left-bottom
ax3 = plt.subplot(2, 2, 3)  # 3rd position
ax3.bar(["A", "B", "C"], [3, 7, 5], color='g')
ax3.set_title("Position (2,2,3)")

# Position 4: Right-bottom
ax4 = plt.subplot(2, 2, 4)  # 4th position
ax4.hist(np.random.randn(1000), bins=30, color='orange', edgecolor='black')
ax4.set_title("Position (2,2,4)")

plt.tight_layout()
plt.show()
```

**Basic Subplot Structure**

`plt.subplot(nrows, ncols, index)`

---


* nrows: Number of rows in the grid
* ncols: Number of columns in the grid
* index: Position of the subplot (starts from 1, moves row-wise)
