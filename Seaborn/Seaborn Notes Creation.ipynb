## Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# Set Seaborn style
sns.set_style("whitegrid")  # Adds a grid background for better readability

# Create the scatter plot
plt.figure(figsize=(8, 5))  # Set figure size for better visualization
sns.scatterplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="day",  # Color points based on the day of the week
    style="time",  # Different marker styles for Lunch and Dinner
    size="size",  # Vary marker sizes based on party size
    #sizes=(20, 200),        # Define size range for points
    palette="deep",  # Use a deep color palette for better distinction
    #color='r'           #for a unique color
    alpha=0.7               # Set transparency
)

# Customization
plt.title("Scatter Plot of Tips Dataset", fontsize=14, fontweight='bold')  # Set title with custom font size and weight
plt.xlabel("Total Bill ($)", fontsize=12)  # Label for x-axis with custom font size
plt.ylabel("Tip Amount ($)", fontsize=12)  # Label for y-axis with custom font size
plt.legend(title="Legend", loc='upper left')  # Position legend in the upper left corner
plt.xticks(fontsize=10)  # Customize x-axis tick font size
plt.yticks(fontsize=10)  # Customize y-axis tick font size

# Show plot
plt.show()  # Display the scatter plot
```

## Rug Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# Set Seaborn style
sns.set_style("whitegrid")  # Adds a grid background for better readability

# Create the rug plot
plt.figure(figsize=(8, 5))  # Set figure size for better visualization
sns.rugplot(
    x="total_bill",               #y=tips['total_bill']  for changing the orientation
    data=tips,
    hue="day",  # Color rugs based on the day of the week
    height=0.1,  # Set the height of the rug markers
    palette="deep",  # Use a deep color palette for better distinction
    #color ='g'       for a unique color
    linewidth=1.5  # Customize the thickness of the rug lines
)

# Customization
plt.title("Rug Plot of Total Bill in Tips Dataset", fontsize=14, fontweight='bold')  # Set title with custom font size and weight
plt.xlabel("Total Bill ($)", fontsize=12)  # Label for x-axis with custom font size
plt.ylabel("Density Indicator", fontsize=12)  # Label for y-axis with custom font size
plt.xticks(fontsize=10)  # Customize x-axis tick font size
plt.yticks([])  # Hide y-axis ticks since rug plots usually don't need them

# Show plot
plt.show()  # Display the rug plot
```

```python
sns.histplot(tips['total_bill'])
sns.rugplot(tips['total_bill'], color='r')    #Overlaying a Rug Plot with a Histogram
plt.title('Histogram with Rug Plot for Total Bill')
plt.show()
```

## Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the built-in 'fmri' dataset from Seaborn
fmri=sns.load_dataset('fmri')
# Create a line plot using Seaborn
sns.lineplot(data=fmri,
             x="timepoint",
             y="signal",
             style="event",  # style=Different line styles for each event type
             markers=True,    # markers=Add markers at data points
             hue="region",         # hue:- for grouping
             palette="Set2") #palette= Use a predefined color palette, #color= Different colors

plt.title(" customizied line plot") # title
plt.xlabel("timepoint") # xlabel
plt.ylabel("signal") # ylabel
plt.show() # show
```

## KDE

```python
import matplotlib.pyplot as plt
import seaborn as sns
# load dataset
tips=sns.load_dataset('tips')
# create kde plot
sns.kdeplot(data=tips,
            x='total_bill',
            hue='sex',  #hue:-groping
            common_norm=False,  # common_norm=False, each groups KDE curve is scaled separately
            fill=True,# fill=True(fill area coverd by curve)
            palette="deep",  # Use a deep color palette for better distinction
            linewidth=2 ) # Set line width for better visibility

plt.title('KDE Plot of Total Bill') # title
plt.xlabel('Total Bill') # x label
plt.ylabel('Density') # y label
plt.show() # show
```

```python
sns.kdeplot(y=tips['total_bill']) # Horizontal kde plot
plt.show()
```

```python
# Bivariate Kdeplot for two for two variables
sns.kdeplot(x=tips['size'],y=tips['total_bill'],color='green',fill=True)
plt.show()
```

## Pair

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# Set Seaborn style
sns.set_style("whitegrid")  # Adds a grid background for better readability

# Create the pair plot
sns.pairplot(
    tips,
    hue="day",  # Color plots based on the day of the week
    palette="deep",  # Use a deep color palette for better distinction
    markers=["o", "s", "D", "^"],  # Set different markers for different categories
    diag_kind="kde",  # Use KDE for diagonal plots instead of histograms   #diag_kind='hist'
    kind="reg"        # add regression line to scatterplots.
)

# Customization
plt.suptitle("Pair Plot of Tips Dataset", fontsize=14, fontweight='bold', y=1.02)  # Set title with custom font size and adjust spacing

# Show plot
plt.show()  # Display the pair plot
```

```python
# Pairplot with only the lower triangle (corner plot)
sns.pairplot(tips, corner=True, height = 5, aspect= 2)              #changing dimention of each subplots
```

## Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# Set Seaborn style
sns.set_style("whitegrid")  # Adds a grid background for better readability

# Create the histogram
plt.figure(figsize=(8, 5))  # Set figure size for better visualization
sns.histplot(
    x="total_bill",
    data=tips,
    hue="day",  # Color bars based on the day of the week
    multiple="stack",  # Stack bars for different categories #multiple='stack'/'dodge'/'layer'
    palette="deep",  # Use a deep color palette for better distinction
    bins=20,  # Set the number of bins for better granularity
    #color='g',       for a unique color
    alpha=0.8,        #transparency
    kde=True,         # Add a kernel density estimate (KDE) curve
    edgecolor="black"  # Add black edges to bars for better clarity
)

# Customization
plt.title("Histogram of Total Bill in Tips Dataset", fontsize=14, fontweight='bold')  # Set title with custom font size and weight
plt.xlabel("Total Bill ($)", fontsize=12)  # Label for x-axis with custom font size
plt.ylabel("Frequency", fontsize=12)  # Label for y-axis with custom font size
plt.legend(title="Legend", loc='upper right')  # Position legend in the upper right corner
plt.xticks(fontsize=10)  # Customize x-axis tick font size
plt.yticks(fontsize=10)  # Customize y-axis tick font size

# Show plot
plt.show()  # Display the histogram
```

## Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# Set Seaborn style
sns.set_style("whitegrid")  # Adds a grid background for better readability

# Create the count plot
plt.figure(figsize=(8, 5))  # Set figure size for better visualization
sns.countplot(
    x="day", #y='day' for horizontal countplot
    data=tips,
    hue="sex",  # Color bars based on gender
    palette="deep",  # Use a deep color palette for better distinction    #color="salmon" for unique color
    linewidth=2,
    order=['Sun', 'Sat', 'Thur', 'Fri'],        # custome order of categories
    edgecolor="black"  # Add black edges to bars for better clarity
)

# Customization
plt.title("Count Plot of Days in Tips Dataset", fontsize=14, fontweight='bold')  # Set title with custom font size and weight
plt.xlabel("Day of the Week", fontsize=12)  # Label for x-axis with custom font size
plt.ylabel("Count", fontsize=12)  # Label for y-axis with custom font size
plt.legend(title="Gender", loc='upper right')  # Position legend in the upper right corner
plt.xticks(fontsize=10,rotation=45)  # Customize x-axis tick font size and rotate x axis label
plt.yticks(fontsize=10)  # Customize y-axis tick font size

# Show plot
plt.show()  # Display the count plot
```

## Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# Set Seaborn style
sns.set_style("whitegrid")  # Adds a grid background for better readability

# Create the violin plot
plt.figure(figsize=(8, 5))  # Set figure size for better visualization
sns.violinplot(
    x="day",
    y="total_bill",
    data=tips,
    hue="sex",  # Color violins based on gender
    palette="deep",  # Use a deep color palette for better distinction
    split=True,  # Split violins to show distribution for each category
    inner="quart",  # Show quartile lines inside the violins
    linewidth=1.2  # Set line width for better clarity
)

# Customization
plt.title("Violin Plot of Total Bill by Day in Tips Dataset", fontsize=14, fontweight='bold')  # Set title with custom font size and weight
plt.xlabel("Day of the Week", fontsize=12)  # Label for x-axis with custom font size
plt.ylabel("Total Bill ($)", fontsize=12)  # Label for y-axis with custom font size
plt.legend(title="Gender", loc='upper right')  # Position legend in the upper right corner
plt.xticks(fontsize=10)  # Customize x-axis tick font size
plt.yticks(fontsize=10)  # Customize y-axis tick font size

# Show plot
plt.show()  # Display the violin plot
```

## Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# Set Seaborn style
sns.set_style("whitegrid")  # Adds a grid background for better readability

# Create the box plot with customizations
plt.figure(figsize=(8, 5))  # Set figure size for better visualization
sns.boxplot(
    x="day",
    y="total_bill",
    data=tips,
    hue="sex",  # Color boxes based on gender
    palette="deep",  # Use a deep color palette for better distinction
    dodge=True,  # Separate boxes for different categories
    #showfliers=False  for removing outliers
    width=0.6,  # Adjust width of boxes
    linewidth=1.2  # Set line width for better clarity
)
# # categorical variable on y axis and numerical on x axis for horizontal boxplots

# Customization
plt.title("Box Plot of Total Bill by Day in Tips Dataset", fontsize=14, fontweight='bold')  # Set title with custom font size and weight
plt.xlabel("Day of the Week", fontsize=12)  # Label for x-axis with custom font size
plt.ylabel("Total Bill ($)", fontsize=12)  # Label for y-axis with custom font size
plt.legend(title="Gender", loc='upper right')  # Position legend in the upper right corner
plt.xticks(fontsize=10)  # Customize x-axis tick font size
plt.yticks(fontsize=10)  # Customize y-axis tick font size

# Show plot
plt.show()  # Display the box plot
```
