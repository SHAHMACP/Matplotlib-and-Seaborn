# penguins

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Penguins dataset
penguins = sns.load_dataset("penguins")

# Set a general style for plots
sns.set_style("whitegrid")

```

```python
# Task 1: Line plot of bill length for each species with hue
plt.figure(figsize=(10, 5))
sns.lineplot(data=penguins, x=penguins.index, y='bill_length_mm', hue='species', style='species', markers=True, dashes=False)
plt.title("Bill Length for Each Species", fontweight='bold')
plt.xlabel("Index")
plt.ylabel("Bill Length (mm)")
plt.legend()
plt.xticks(rotation=45)
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Penguins dataset
penguins = sns.load_dataset("penguins")

# Drop non-numerical columns
penguins_numeric = penguins.select_dtypes(include=["float64", "int64"])

# Compute the correlation matrix
corr_matrix = penguins_numeric.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,  # Display correlation values
    fmt=".2f",  # Format to 2 decimal places
    cmap="coolwarm",  # Use an appropriate colormap
    linewidths=0.5,  # Add spacing between cells
    linecolor="black",
    cbar=True
)

# Customize the plot
plt.title("Correlation Heatmap of Numerical Features in Penguins Dataset", fontsize=14, fontweight="bold")
plt.xlabel("Features", fontsize=12)
plt.ylabel("Features", fontsize=12)

# Show the plot
plt.show()
```

```python
# Task 3: Bar chart of average flipper length per species with hue
plt.figure(figsize=(8, 5))
sns.barplot(data=penguins, x='species', y='flipper_length_mm', hue='species')
plt.axhline(penguins['flipper_length_mm'].mean(), color='red', linestyle='dashed', label='Overall Mean')
plt.legend()
plt.title("Average Flipper Length per Species")
plt.ylabel("Flipper Length (mm)")
plt.show()

```

```python
# Task 4: Grouped bar chart for bill depth by gender and species with hue
plt.figure(figsize=(8, 6))
sns.barplot(data=penguins, x='species', y='bill_depth_mm', hue='sex', ci=None)
plt.title("Average Bill Depth by Gender and Species")
plt.ylabel("Bill Depth (mm)")
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Sex", loc='upper right')
plt.show()

```

```python
# Task 5: Histogram of body mass with density curve and hue
plt.figure(figsize=(8, 5))
sns.histplot(data=penguins, x='body_mass_g', hue='species', bins=20, kde=True, edgecolor='black', alpha=0.7)
plt.axvline(penguins['body_mass_g'].median(), color='red', linestyle='dashed', label='Median')
plt.legend()
plt.title("Histogram of Body Mass")
plt.xlabel("Body Mass (g)")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Penguins dataset
penguins = sns.load_dataset("penguins")

# Create a pairplot
sns.pairplot(
    penguins,
    hue="species",  # Color by species
    markers=["o", "s", "D"],  # Different markers for each species
    diag_kind="kde",  # Show density plots on the diagonal
    height=2.5  # Adjust figure size
)

# Add a title
plt.suptitle("Pairplot of Numerical Features in Penguins Dataset", fontsize=14, fontweight="bold", y=1.02)

# Show the plot
plt.show()


```

```python
# Task 7: Scatter plot of flipper length vs. body mass with hue
plt.figure(figsize=(8, 6))
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='species', size='species', alpha=0.7)
plt.title("Flipper Length vs. Body Mass")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.show()

```

```python
# Create a boxplot for flipper length by species
plt.figure(figsize=(8, 6))
sns.boxplot(data=penguins, x="species", y="flipper_length_mm", palette="Set2", width=0.6)

# Customize plot
plt.title("Boxplot of Flipper Length by Species", fontweight='bold')
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Penguins dataset
penguins = sns.load_dataset("penguins")

# Create the count plot
plt.figure(figsize=(8, 6))
sns.countplot(
    data=penguins,
    x="species",
    hue="sex",  # Different colors for male and female
    palette={"Male": "#1f77b4", "Female": "#ff7f0e"},  # Custom colors
    edgecolor="black",  # Customize bar edge color
    linewidth=1.5  # Customize bar edge width
)

# Customize the plot
plt.xlabel("Species", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Count of Male and Female Penguins by Species", fontsize=14, fontweight="bold")

# Rotate x-axis labels
plt.xticks(rotation=20)

# Modify legend position
plt.legend(title="Sex", loc="lower center", bbox_to_anchor=(0.5, -0.15), ncol=2)

# Show the plot
plt.show()
```

# Titanic

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Set style
sns.set_style("whitegrid")



```

```python
# 1. Line Plot: Age of Passengers for Each Class
plt.figure(figsize=(8,6))
sns.lineplot(data=titanic, x=titanic.index, y="age", hue="pclass", marker="o", linestyle="-")
plt.title("Age of Passengers by Class", fontsize=14, fontweight='bold')
plt.xlabel("Passenger Index")
plt.ylabel("Age")
plt.legend(title="Class")
plt.show()
```

```python
# 2

import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Select relevant numerical features
titanic_numeric = titanic[["survived", "pclass", "age", "fare", "sibsp", "parch"]].dropna()

# Create the pairplot
sns.pairplot(
    titanic_numeric,
    hue="survived",  # Color based on survival status
    markers=["o", "s"],  # Customize marker styles for survivors and non-survivors
    diag_kind="kde",  # Show distribution as density plots
    palette={0: "red", 1: "green"},  # Red for non-survivors, Green for survivors
    height=2.5  # Adjust figure size
)

# Add a title
plt.suptitle("Pairplot of Numerical Features in Titanic Dataset", fontsize=14, fontweight="bold", y=1.02)

# Show the plot
plt.show()
```

```python
# 3. Histogram: Age Distribution
plt.figure(figsize=(8,6))
sns.histplot(titanic["age"], bins=20, kde=True, color='blue', edgecolor='black', alpha=0.6)
plt.title("Age Distribution of Passengers", fontsize=14, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python

#4
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Select only numerical columns
titanic_numeric = titanic.select_dtypes(include=["float64", "int64"])

# Compute the correlation matrix
corr_matrix = titanic_numeric.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,  # Display correlation values
    fmt=".2f",  # Format to 2 decimal places
    cmap="coolwarm",  # Use an appropriate colormap
    linewidths=0.5,  # Add spacing between cells
    linecolor="black",
    cbar=True
)

# Customize the plot
plt.title("Correlation Heatmap of Numerical Features in Titanic Dataset", fontsize=14, fontweight="bold")
plt.xlabel("Features", fontsize=12)
plt.ylabel("Features", fontsize=12)

# Show the plot
plt.show()
```

```python
# 5. Scatter Plot: Age vs Fare by Survival
plt.figure(figsize=(8,6))
sns.scatterplot(data=titanic, x="age", y="fare", hue="survived", style="pclass", size="pclass",
                palette={0:"red", 1:"green"}, sizes=(20,100))
plt.title("Age vs Fare Colored by Survival", fontsize=14, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# 6. Boxplot: Age Distribution by Class
plt.figure(figsize=(8,6))
sns.boxplot(data=titanic, x="pclass", y="age", palette="pastel", width=0.6)
plt.title("Age Distribution by Class", fontsize=14, fontweight='bold')
plt.xlabel("Class")
plt.ylabel("Age")
plt.show()
```

```python
# 7. Count Plot: Passenger Count by Class & Survival
plt.figure(figsize=(8,6))
sns.countplot(data=titanic, x="pclass", hue="survived", palette="coolwarm")
plt.title("Passenger Count by Class and Survival", fontsize=14, fontweight='bold')
plt.xlabel("Class")
plt.ylabel("Count")
plt.legend(title="Survived")
plt.show()
```

```python
#8
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Drop missing values in the age column
titanic = titanic.dropna(subset=["age"])

# Create the KDE plot
plt.figure(figsize=(8, 5))
sns.kdeplot(
    data=titanic,
    x="age",
    hue="survived",  # Different colors for survivors and non-survivors
    fill=True,  # Fill the area under the curve
    alpha=0.5,  # Adjust transparency
    linestyle="--",  # Customize line style
    palette={0: "red", 1: "green"}  # Red for non-survivors, Green for survivors
)

# Customize the plot
plt.xlabel("Age", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.title("Age Distribution of Titanic Passengers by Survival Status", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)  # Add grid lines

# Show the plot
plt.show()
```

```python
# 9. Grouped Bar Chart: Survival by Gender and Class
plt.figure(figsize=(8,6))
sns.barplot(data=titanic, x="pclass", y="survived", hue="sex", ci=None, palette="coolwarm")
plt.title("Survival Rate by Gender and Class", fontsize=14, fontweight='bold')
plt.xlabel("Class")
plt.ylabel("Survival Rate")
plt.legend(title="Gender", loc='upper right')
plt.grid(axis='y', linestyle='dashed', alpha=0.7)
plt.show()

```

# Diomond

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the Diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Set style
sns.set_style("whitegrid")


```

```python
# 1. Line Plot: Price of Diamonds by Cut
plt.figure(figsize=(8,6))
sns.lineplot(data=diamonds, x=diamonds.index, y="price", hue="cut", marker="o", linestyle="-")
plt.title("Price of Diamonds by Cut", fontsize=14, fontweight='bold')
plt.xlabel("Index")
plt.ylabel("Price")
plt.legend(title="Cut")
plt.show()
```

```python
# 2. Line Plot: Carat Variation by Clarity
plt.figure(figsize=(8,6))
sns.lineplot(data=diamonds, x=diamonds.index, y="carat", hue="clarity", marker="s", linestyle="--")
plt.title("Carat Variation by Clarity", fontsize=14, fontweight='bold')
plt.xlabel("Index")
plt.ylabel("Carat")
plt.legend(title="Clarity")
plt.grid(True)
plt.show()
```

```python
# 3. Bar Chart: Average Depth by Cut
plt.figure(figsize=(8,6))
sns.barplot(data=diamonds, x="cut", y="depth", ci=None, palette="muted")
plt.title("Average Depth by Cut", fontsize=14, fontweight='bold')
plt.xlabel("Cut")
plt.ylabel("Average Depth")
plt.legend()
plt.show()
```

```python
# 4. Grouped Bar Chart: Table Size by Color
plt.figure(figsize=(8,6))
sns.barplot(data=diamonds, x="color", y="table", hue="cut", ci=None, palette="coolwarm")
plt.title("Table Size by Diamond Color", fontsize=14, fontweight='bold')
plt.xlabel("Color")
plt.ylabel("Table Size")
plt.legend(title="Cut", loc='upper right')
plt.grid(axis='y', linestyle='dashed', alpha=0.7)
plt.show()
```

```python
# 5. Histogram: Price Distribution
plt.figure(figsize=(8,6))
sns.histplot(diamonds["price"], bins=30, kde=True, color='blue', edgecolor='black', alpha=0.6)
plt.axvline(diamonds["price"].median(), color='red', linestyle='dashed', label=f"Median: {diamonds['price'].median()}")
plt.title("Price Distribution of Diamonds", fontsize=14, fontweight='bold')
plt.xlabel("Price")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# 7. Scatter Plot: Carat vs Price by Cut
plt.figure(figsize=(8,6))
sns.scatterplot(data=diamonds, x="carat", y="price", hue="cut", style="clarity", size="clarity",
                palette="coolwarm", sizes=(20,100))
plt.title("Carat vs Price by Cut", fontsize=14, fontweight='bold')
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show()
```

```python
# 8. Boxplot: Price Distribution by Cut
plt.figure(figsize=(8,6))
sns.boxplot(data=diamonds, x="cut", y="price", palette="pastel", width=0.6)
plt.title("Price Distribution by Cut", fontsize=14, fontweight='bold')
plt.xlabel("Cut")
plt.ylabel("Price")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create the KDE plot
plt.figure(figsize=(8, 5))
sns.kdeplot(
    data=diamonds,
    x="price",
    hue="cut",  # Different colors for different cut qualities
    fill=True,  # Fill the area under the curve
    alpha=0.5,  # Adjust transparency
    linestyle="--"  # Customize line style
)

# Customize the plot
plt.xlabel("Price", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.title("Price Distribution of Diamonds by Cut Quality", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)  # Add grid lines

# Show the plot
plt.show()
```
