```python
# 2. Multiple Line Plots
plt.figure(figsize=(6,4))
plt.plot(months, product_A_sales, 'r-o', label='Product A')
plt.plot(months, product_B_sales, 'b-s', label='Product B')
plt.plot(months, product_C_sales, 'g-d', label='Product C')
plt.title('Sales Trends of Products', fontsize=20, fontfamily='serif', color='blue')
plt.legend(loc='lower right')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
# 3. Bar Chart
plt.figure(figsize=(6,4))
plt.bar(['Product A', 'Product B', 'Product C'], [900,850,70], color=['red', 'blue', 'green'])
plt.xticks(rotation=40)
plt.title('Sales Comparison in May')
plt.ylabel('Sales')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
steps = [5000, 7000, 8000, 10000, 12000]  # Steps count
calories_burned = [200, 300, 400, 500, 600]  # Calories burned
workout_duration = [30, 40, 50, 60, 75]  # Workout duration in minutes

# Create a 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# (1) Line Plot for Steps
axes[0, 0].plot(months, steps, marker='o', linestyle='-', color='blue', label="Steps")
axes[0, 0].set_title("Steps Trend Over 5 Months")
axes[0, 0].set_xlabel("Months")
axes[0, 0].set_ylabel("Steps Count")
axes[0, 0].grid(True, linestyle='--', alpha=0.5)
axes[0, 0].legend()

# (2) Bar Chart Comparing Fitness Parameters in May
fitness_params = ['Steps', 'Calories Burned', 'Workout Duration']
may_values = [steps[-1], calories_burned[-1], workout_duration[-1]]
axes[0, 1].bar(fitness_params, may_values, color=['blue', 'green', 'red'])
axes[0, 1].set_title("Comparison of Fitness Parameters in May")
axes[0, 1].set_ylabel("Values")
axes[0, 1].tick_params(axis='x', rotation=40)

# (3) Pie Chart of Total Fitness Contribution Over 5 Months
total_values = [sum(steps), sum(calories_burned), sum(workout_duration)]
explode = [0.1 if val == max(total_values) else 0 for val in total_values]  # Explode highest value
axes[1, 0].pie(total_values, labels=fitness_params, autopct='%1.1f%%', explode=explode,
               colors=['blue', 'green', 'red'], shadow=True, startangle=140)
axes[1, 0].set_title("Total Contribution of Fitness Parameters")

# (4) Scatter Plot of Workout Duration vs. Calories Burned
axes[1, 1].scatter(workout_duration, calories_burned, s=[w * 2 for w in workout_duration],
                   c=['blue', 'green', 'red', 'purple', 'orange'], alpha=0.8, edgecolors='black')
axes[1, 1].set_title("Workout Duration vs. Calories Burned")
axes[1, 1].set_xlabel("Workout Duration (min)")
axes[1, 1].set_ylabel("Calories Burned")
axes[1, 1].grid(True, linestyle='--', alpha=0.5)

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig("fitness_analysis.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
temperature = [15, 18, 22, 25, 28]  # in Celsius
rainfall = [80, 60, 50, 30, 20]  # in mm
humidity = [70, 65, 60, 55, 50]  # in percentage

# Convert data to a DataFrame
df = pd.DataFrame({
    'Temperature (C)': temperature,
    'Rainfall (mm)': rainfall,
    'Humidity (%)': humidity
}, index=months)

# Prepare data for pie chart
labels = ['Temperature (C)', 'Rainfall (mm)', 'Humidity (%)']
values = [sum(temperature), sum(rainfall), sum(humidity)]
explode = [0.1 if value == max(values) else 0 for value in values]  # Explode highest value

# Create 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Line Plot for Temperature
axes[0, 0].plot(months, temperature, linestyle='dotted', marker='o', color='orange', markersize=8)
axes[0, 0].set_title("Temperature Trend Over 5 Months", fontsize=12)
axes[0, 0].set_xlabel("Months")
axes[0, 0].set_ylabel("Temperature (C)")

# 2. Bar Chart Comparing Weather Parameters in May
weather_may = [temperature[-1], rainfall[-1], humidity[-1]]
axes[0, 1].bar(labels, weather_may, color=['red', 'blue', 'green'])
axes[0, 1].set_title("Weather Parameters in May", fontsize=12)
axes[0, 1].set_xlabel("Weather Parameters")
axes[0, 1].set_ylabel("Values")

# 3. Pie Chart of Total Weather Contribution
axes[1, 0].pie(values, labels=labels, autopct='%1.1f%%', explode=explode, colors=['red', 'blue', 'green'],
               startangle=140, shadow=True, wedgeprops={'edgecolor': 'black'})
axes[1, 0].set_title("Total Weather Contribution Over 5 Months", fontsize=12)

# 4. Scatter Plot of Temperature vs. Rainfall
axes[1, 1].scatter(temperature, rainfall, color='purple', alpha=0.8, s=100)
axes[1, 1].set_title("Temperature vs. Rainfall", fontsize=12)
axes[1, 1].set_xlabel("Temperature (C)")
axes[1, 1].set_ylabel("Rainfall (mm)")

# Adjust layout
plt.tight_layout()

# Save the figure as a PDF
plt.savefig("weather_analysis.pdf", format='pdf')

# Show plot
plt.show()
```
